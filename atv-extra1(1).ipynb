```python
import math 
def gerador_quantizacao():
    n_niveis = int(input("Digite o número de níveis de quantização: "))

    n_bits = math.ceil(math.log2(n_niveis))

    print(f"\n--- Resultado ---")
    print(f"Niveis desejados: {n_niveis}")
    print(f"Bits necessários: {n_bits}")
    print(f"Capacidade real com esses bits: {2**n_bits} combinações") 
    print(f"---------------------\n")


    print(f"Combinações de {n_bits} bits: ")

    for i in range(2**n_bits): 
        combinacao = f"{i:0{n_bits}b}"
        print (f"Nivel {i:3}: {combinacao}")

if __name__ == "__main__": 
    gerador_quantizacao() 
```

```python
def decimal_binario(n): 
    if (n) == 0:
        return "0"

    bits = ""
    while n > 0: 
        resto = n % 2
        bits = str(resto) + bits
        n = n // 2
    
    return bits

numero = int(input("Digite um número décimal:")) 
print (f"O binario de {numero} é: {decimal_binario(numero)}") 
```

```python
def decimal_binario_fracao(n, precisao=10):
    inteira=int(n)
    fracionaria = n - inteira

    bin_inteira = bin(inteira).replace("0b", "")

    bin_frac = [] 
    for _ in range(precisao):
        fracionaria *= 2
        bit = int(fracionaria)
        bin_frac.append(str(bit))
        fracionaria -= bit
        if fracionaria == 0: break

    return f"{bin_inteira}.{''.join(bin_frac)}"

def binario_decimal_fracao(bin_str):
    if "." not in bin_str: 
        return float(int(bin_str, 2))

    inteira_str, frac_str = bin_str.split(".")
    
    decimal_inteira = int(inteira_str, 2)
    decimal_frac = 0

    for i, bit in enumerate (frac_str):
        if bit == "1":
            decimal_frac += 2**-(i + 1)

    return decimal_inteira + decimal_frac

print("Selecione a Operação: ")
print(" 1 - Decimal para Binário")
print(" 2 - Binário para Decimal")
opcao = input("Opção (1 ou 2): ")

if opcao == "1":
    num = float(input("Digite o número décimal (ex: 10.324): "))
    print (f"Resultado em Binário {decimal_binario_fracao(num)}")

elif opcao == "2":
    b_str = input("Digite o binário (ex:1010.101): ")
    print(f"Resultado em binário: {binario_decimal_fracao(b_str)}") 

else:
    print("Opção Inválida") 
```

```python
def hex_bin(h):
    return bin(int(h, 16)) [2:]

def octal_bin(o):
    return bin(int(o, 8)) [2:]

def bin_decimal(b):
    decimal = 0
    bits_invertidos = b [::-1] 
    for i, bit in enumerate (bits_invertidos):
        if bit == "1":
            decimal += 2**i
    return decimal

print ("Convensor de bases (via binário)") 
print ("1 - Hexadecimal -> Binário -> Decimal") 
print ("2 - Octal -> Binário -> Decimal")
opcao = input("Escolha (1 ou 2): ") 

if opcao == "1":
    entrada = input("Digite o valor Hexa (ex: 1A): ")
    binario = hex_bin(entrada)
    decimal = bin_decimal(binario)
    print(f"\nHexa: {entrada} -> Binário: {binario} -> Decimal: {decimal}")


elif opcao == "2":
    entrada = input("Digite o valor Octal (ex: 15): ")
    binario = octal_bin(entrada)
    decimal = bin_decimal(binario)
    print(f"\nOctal: {entrada} -> Binário: {binario} -> Decimal: {decimal}")

else:
    print("Opção Inválida") 
```
